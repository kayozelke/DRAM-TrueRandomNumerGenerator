```python
import matplotlib.pyplot as plt

numbers = []
numbers_8bit = []

# Wczytanie liczb z pliku
#with open('plik16.txt') as f:
with open('probki_102400.txt') as f:
    for x in f.readlines():
        x_int = int(x)  # konwersja na liczbę całkowitą
        numbers.append(int(x_int))


with open('wstepne_przetwarzanie.txt', 'w') as f:
    for x in numbers:
        bits = format(x & 0b111, '03b')  # pobieramy 3 ostatnie bity i formatujemy jako 3-znakowy ciąg binarny z wiodącymi zerami
        f.write(bits)  # zapisujemy do pliku

with open('wstepne_przetwarzanie.txt') as f:
    line = f.readline().strip()  # odczytujemy jedyną linię pliku i usuwamy białe znaki na końcach

    # tworzymy nową listę z podziałem każdych 8 znaków na osobne elementy
    data = [line[i:i+8] for i in range(0, len(line), 8)]
    for string_x in data:
        # zamiana ciągu znaków string
        # "2" w drugim argumencie oznacza konwersję z systemu binarnego
        numbers_8bit.append(int(string_x, 2))

        
print("Wartosci probek wprost ze zrodla: ", numbers)
print( "Liczba probek: " + str(len(numbers)) + ", max próbka: " + str(max(numbers)) + ", min próbka: " + str(min(numbers)))

print("Liczby binarne po zebraniu ostatnich N bitow i złożeniu do paczki odczytywanej jako liczby 8-bitowe: ", data)
print("Te same liczby w postaci decymalnej: ", numbers_8bit)

```

```python
import numpy as np
import math

def entropy1(labels, base=None):
  value,counts = np.unique(labels, return_counts=True)
  norm_counts = counts / counts.sum()
  base = 2 if base is None else base
  return -(norm_counts * np.log(norm_counts)/np.log(base)).sum()


print("Entropia źródła: ",entropy1(numbers, 2))

print("Entropia po zebraniu ostatnich N bitow i złożeniu do paczki odczytywanej jako liczby 8-bitowe:", entropy1(numbers_8bit, 2))
#print(calculate_entropy(numbers_8bit))
```

```python
# Wyświetlenie histogramu
plt.figure(figsize=(15, 4))
y,x, _ = plt.hist(numbers_8bit, bins=256, edgecolor='black')
plt.xlabel('Wartość(x)')
plt.ylabel('Liczba wystąpień')
plt.title('Histogram')
plt.xlim(0,256)
plt.show()


print("Wartość maksymalna dla: ", str(x[np.where(y == y.max())]), ", wynosi: ", y.max())
print("Wartość minimalna dla: ", str(x[np.where(y == y.min())]), ", wynosi: ", y.min())
print("Średnia ilość wystąpień: ", sum(y)/len(y), " przy ", len(numbers_8bit), " 8-bitowych liczb")
```

```python
from Crypto.Cipher import AES



key = b'Klucz Szyfrujacy'
cipher = AES.new(key, AES.MODE_ECB)
encrypted_16bit_packs = []

with open('wstepne_przetwarzanie.txt', 'rb') as f:
    while True:
        # czytaj 16 bitów
        data = f.read(16)
        
        # jeśli nie udało się nic więcej odczytać, przerwij pętlę
        if not data:
            break
    

        encrypted_16bit = cipher.encrypt(data)
        encrypted_16bit_packs.append(encrypted_16bit)


print("Klucz szyfrujacy: ", key)
print("Ostatnia paczka zaszyfrowana 16bit: ", encrypted_16bit)
print("Liczba wszystkich paczek zaszyfrowanych 16bit:", len(encrypted_16bit_packs))

# Konwertowanie zaszyfrowanych danych na ciąg zer i jedynek
binary_data = ''
for pack in encrypted_16bit_packs:
    for b in pack:
        binary_data += format(b, '08b')
# Zapisanie ciągu zer i jedynek do pliku
with open('encrypted_16bit.txt', 'w') as f:
    f.write(binary_data)
```
